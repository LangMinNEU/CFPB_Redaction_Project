## 1. Setup

```python
import pandas as pd
```

## 2. Load file

```python
df_results = pd.read_csv('result_0925.csv')  # Use the result file <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
print("File read successfully.")
print(f'Number of complaints before filtering: {df_results.shape[0]}')
print()

df = df_results.sample(n=100, random_state=42)
print(f'Number of complaints after filtering: {df.shape[0]}')
print()
```

## 3. Output

```python
output_file = "result_0925_sample.csv"  # Choose the name for output <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

new_file = pd.DataFrame(columns=['Complaint ID', 'Date', 'Product', 'Issue', 'Complaint', 'Answer', 'Explanation', 'Label', 'Category'])
new_file.to_csv(output_file, index=False)
df.to_csv(output_file, mode='a', header=False, index=False)
print("File output successfully.")
```
